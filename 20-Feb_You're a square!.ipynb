```python
def is_square(n):
    try:
        if n>=0 and round((n)**(1/2), 0) == ((n)**(1/2)):
            return True
        else:
            return False
    except TypeError:
        print("What are you trying, bro?")
```

```python
is_square(25)
```

```python
is_square((-1))
```

```python
is_square(('a'))
```
